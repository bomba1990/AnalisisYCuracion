```python
tabla_2 = pd.read_json(BASE_DIR + "/input/pasos.json")
print(tabla_2.shape)
tabla_2.head()
```

```python
tabla_2.province.unique()
```

```python
tabla_2.replace('C√≥rdoba','Cordoba',inplace=True)
tabla_2.replace('AEROPUERTO INTERNACIONAL PILOTO CIVIL NORBERTO FERNANDEZ','Santa Cruz',inplace=True)
tabla_2.replace('NEUQUEN','Neuquen',inplace=True)
```

```python
plt.figure(figsize=(10,10))
ax = sns.countplot(y='province',data=tabla_2,order=tabla_2["province"].value_counts().index)
ax.set_ylabel("Provincia")
ax.set_title("Cantidad de Pasos Fronterizos por provincia");
```
